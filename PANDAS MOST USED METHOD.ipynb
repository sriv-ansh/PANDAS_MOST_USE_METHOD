### 21.handling missing data

We can Handle the Missing Value by droping or by using the filling missing value technique. As per companies Standard if Our Data of Columns Containts More Then 33% missing Value Then We have the drop the entire columns Else We Need to fill the Missing value using some technique 

### 22 isnull() / notnull()

```python
### This is percentage of Missing Value in Our Columns
(df.isnull().sum() * 100 / len(df)).round(2)
```

### Notnull

```python
(df.notnull().sum() * 100 // len(df)).round(4)
```

### 23.dropping missing values


```python
#### As We can See Our Deck Columns Contain More the 77% missing Values So We Can directly Drop the column
df.drop('Deck', axis =1 , inplace = True)
df
```

### 24.filling missing values

```python
df.isnull().sum() / len(df) * 100
```

```python
### We Can Missing the missing age values using Mean of Age
mean_age = df['Age'].mean()
print(mean_age)

df['Age'].fillna(mean_age).isnull().sum()

### As We Can See Now We Have 0 Missing Value in Age Column But We Haven't Use the inplace , perameter because we are going to 
### use another method of handling the missing values in Age Column
```

```python
df['Age'].fillna(df['Age'].mode().iloc[0],inplace = True)
```

```python
df['Age'].isnull().sum()
```

### 27.combining dataframe--> merging-->join--> inner/left/right/outer

```python
students = pd.DataFrame({'ID':[1,2,3,4],'name':['Andrew','Bobo','Claire','David'],'course_id':[3,2,4,1]})
course = pd.DataFrame({'course_id':[1,2,3,4],'course':['Excel','Python','Mysql','Java']})
```

```python
students
```

```python
course
```

### Join
* Join is used to join the Data in Sql as same we can use it in pandas as well
* There Are 5 types of join Inner , Outer , left , right , cross , inner

### Inner Join

```python
pd.merge(students,course, how = 'inner', on = 'course_id')
```

### Left Outer Join

```python
course['course_id'].replace({4:6},inplace = True)
```

```python
pd.merge(students, course, how = 'left', on = 'course_id')
```

### Right Outer Join

```python
pd.merge(students, course, how = 'right', on = 'course_id')
```

### Cross Join

```python
pd.merge(students, course, how = 'cross')
```
